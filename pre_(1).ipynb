```python
# 选择相邻的列并将它们合并成元组
merged_columns = [data_new.iloc[:, i:i+2].apply(tuple, axis=1) for i in range(0, data_new.shape[1], 2)]

# 创建一个新的DataFrame，每列都是合并后的元组
result_df = pd.concat(merged_columns, axis=1)

#欄位名稱
result_df.columns=[colnames]

#編寫原點函數
def to_origin(p1,p2=(0,0)):
    x, y = p1
    a, b = p2
    return ((x-a)**2 + (y-b)**2)**0.5

#函數帶回每一個columns
origin_df=result_df.copy()
for col in origin_df.columns:
    origin_df[col] = origin_df[col].apply(to_origin)

# 打印结果
# print(result_df.head())
# print(origin_df.head())

# 每個row的最小值發生於哪個columns
min_col=origin_df.idxmin(axis=1)
print(min_col)

# 每個row的最大值發生於哪個columns
max_col=origin_df.idxmax(axis=1)
print(max_col)
```

```python
# 從每個row的最小值發生於哪個columns,回result_df找值
min_val = [result_df.loc[index, col] for index, col in zip(result_df.index, min_col)]
# min_val=[(0,1),(1,2),(2,3)]
print(min_val)

# 從每個row的最大值發生於哪個columns,回result_df找值
max_val = [result_df.loc[index, col] for index, col in zip(result_df.index, max_col)]
# max_val=[(4,5),(5,6),(6,7)]
print(max_val)

#找出每筆row最大最小的距離
l_std=[]
for i in range(len(min_val)):
  a=to_origin(max_val[i],min_val[i])
  l_std.append(a)
print(l_std) #std
```

```python
class csv_std:
    def __init__(self, file, target_row):
        self.file = file
        self.target_row = target_row

    def load(self):
        self.df = pd.read_csv(self.file, header=[1, 2], index_col=0)

    def to_std(self):
        to_drop = [(col[0], col[1]) for col in self.df.columns if col[1] == "likelihood"]
        self.df = self.df.drop(columns=to_drop)
        mean = self.df.mean()
        std = self.df.std()
        self.z_score = (self.df - mean) / std

    def re_rows(self):
        if self.z_score.shape[0] > self.target_row:
            self.z_score = self.z_score.iloc[:self.target_row]
        elif self.z_score.shape[0] < self.target_row:
            padding_rows = self.target_row - self.z_score.shape[0]
            pad_df = pd.DataFrame(0, index=range(padding_rows), columns=self.z_score.columns)
            self.z_score = pd.concat([self.z_score, pad_df], axis=0).reset_index(drop=True)

    def save(self, filename):
        csv_std_path=os.path.join(os.path.dirname(self.file), "csv_std_data")
        if not os.path.exists(csv_std_path):
            os.makedirs(csv_std_path)
        output_path = os.path.join(csv_std_path, filename)
        self.z_score.to_csv(output_path, index=False)

    def auto_run(self, filename):
        self.load()
        self.to_std()
        self.re_rows()
        self.save(filename)
```

```python
for idx, file_name in enumerate(os.listdir(folder_path), 1):
    if file_name.endswith('.csv'):
      file_path = os.path.join(folder_path, file_name)
      std_csv = csv_std(file_path, 140)
      std_csv .auto_run(f"std_csv{str(idx).zfill(2)}.csv")
folder = os.path.join(folder_path, "csv_std_data")
all_files = [os.path.join(folder, file) for file in os.listdir(folder) if file.endswith('.csv')]
first_df = pd.read_csv(all_files[0], header=[0, 1])
list_of_csv = [first_df] + [pd.read_csv(file, header=None, skiprows=2) for file in all_files[1:]]
for df in list_of_csv[1:]:
    df.columns = first_df.columns
merged_df = pd.concat(list_of_csv, ignore_index=True)
merged_df.to_csv(os.path.join(folder, "all_csv_data.csv"), index=False)
```
